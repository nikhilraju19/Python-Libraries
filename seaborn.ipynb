1. Write a program to draw bar plot of sex against survived for a dataset given in the url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")
sns.barplot(x="sex", y="survived", hue="sex", data=df, errorbar=None, palette="pastel", legend=True,width = 0.3)

plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender on Titanic")
plt.show()
```

8. Write a program to draw a swarm plot of total bill against size for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
sns.swarmplot(x="size", y="total_bill", data=df, s = 4)

plt.xlabel("Size of Party")
plt.ylabel("Total Bill Amount")
plt.title("Swarm Plot of Total Bill by Party Size")
plt.show()
```
