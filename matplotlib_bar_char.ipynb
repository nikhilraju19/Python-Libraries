1. Write a Python programming to display a bar chart of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt

programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
plt.bar(programming_languages, popularity, width = 0.3)

plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("opularity of Programming Languages")
plt.show()
```

7. Write a Python programming to display a bar chart of the popularity of programming
Languages. Specify the position of each bar plot.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import numpy as np

programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
plt.bar(programming_languages, popularity, width = 0.3, color=['blue', 'green', 'red', 'purple', 'orange', 'cyan'], edgecolor = "darkblue", linewidth = 1.5)

positions = np.arange(len(programming_languages))
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')
plt.xticks(positions, programming_languages)
plt.ylim(0, max(popularity) + 5)

for i, value in enumerate(popularity):
    plt.text(i, value + 0.5, str(value), ha='center', fontsize=8)

plt.show()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)

```python
groups = ['A','B','C','D', 'E'] 
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

X_axis = np.arange(len(groups)) 

plt.bar(X_axis - 0.1, men, width = 0.2, label='Men', color='blue')
plt.bar(X_axis + 0.1, women, width = 0.2, label='Women', color='red')

plt.xticks(X_axis, groups)
plt.xlabel("Group")
plt.ylabel("Scores")
plt.title("Scores by Group and Gender")
plt.legend()

plt.show()
```

11. Write a Python program to create bar plot from a DataFrame.
Sample Data Frame:
a b c d e
2 4,8,5,7,6
4 2,3,4,2,6
6 4,7,4,7,8
8 2,6,4,8,6
10 2,4,3,3,2

```python
import pandas as pd
data = {
    'a': [2, 4, 6, 8, 10],
    'b': [4, 2, 4, 2, 4],
    'c': [8, 3, 7, 6, 4],
    'd': [5, 4, 4, 4, 3],
    'e': [7, 2, 7, 8, 3],
    'f': [6, 6, 8, 6, 2]
}

df = pd.DataFrame(data)

df.plot(kind='bar', figsize=(8, 6))

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Bar Plot from DataFrame")
plt.legend(title="Columns")
plt.show()
```

12. Write a Python program to create bar plots with error bars on the same figure. Sample Date
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
x_labels = ['A', 'B', 'C', 'D']

x_axis = np.arange(len(x_labels))
plt.bar(x_axis, velocity, yerr=std_dev_velocity, capsize=5, color='blue', alpha=0.7, width = 0.2)

plt.xticks(x_axis, x_labels)
plt.xlabel("Category")
plt.ylabel("Mean Velocity")
plt.title("Bar Plots with Error Bars")
plt.show()
```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).
Sample Data
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
x_labels = ['A', 'B', 'C', 'D']

x_axis = np.arange(len(x_labels))
plt.xticks(x_axis, x_labels)
plt.xlabel("Category")
plt.ylabel("Mean Velocity")
plt.title("Bar Plots with Error Bars")
fig = plt.bar(x_axis, velocity, yerr=std_dev_velocity, capsize=5, color='blue', alpha=0.7, width = 0.2)
for bar in fig:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height + 0.02, f'{height:.2f}', 
                 ha='center', va='bottom', fontsize=8, fontweight='bold', color='green')
plt.show()
```
